Task 4 : Analyzing and visualizing sentiment patterns in social media
data to understand public opinion and attitudes towards
specific topics or brands.

Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading Dataset

```python
df = pd.read_csv('/content/twitter_training.csv', header=None)
df.columns = ['ID', 'Game', 'Sentiment', 'Text']

df.head()
```

Basic Understanding

```python
df.info()
df['Sentiment'].value_counts()
```

Sentiment Distribution (Bar Chart)

```python
df['Sentiment'].value_counts().plot(kind='bar')

plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')

plt.show()
```

Sentiment Percentage (Pie Chart)

```python
df['Sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title('Sentiment Percentage')
plt.ylabel('')

plt.show()
```

Sentiment by Game

```python
sns.countplot(x='Game', hue='Sentiment', data=df)

plt.title('Sentiment by Game')
plt.xticks(rotation=45)

plt.show()
```

Real Example interpretation

```python
for sentiment in df['Sentiment'].unique():
    print(f"\n{sentiment} examples:")
    print(df[df['Sentiment'] == sentiment]['Text'].head(2))
```

Conclusion :

Sentiment analysis was performed on social media text data to understand public opinion towards different games. The dataset contains text along with sentiment labels such as Positive, Negative, and Neutral.
